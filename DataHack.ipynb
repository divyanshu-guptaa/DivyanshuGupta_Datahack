```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X_train = pd.read_csv('training_set_features.csv',index_col='respondent_id')
```

```python
X_test = pd.read_csv('test_set_features.csv',index_col='respondent_id')
y_train = pd.read_csv('training_set_labels.csv',index_col='respondent_id')
```

```python
X_test.info()
```

```python
from sklearn import preprocessing 

categorical_columns = X_train.select_dtypes(include=['object']).columns.tolist()
label_encoder = preprocessing.LabelEncoder() 
for column in categorical_columns:
    X_train[column]= label_encoder.fit_transform(X_train[column]) 
    X_test[column]= label_encoder.fit_transform(X_test[column]) 
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
X_train = pd.DataFrame(imputer.fit_transform(X_train),columns=X_train.columns,index=X_train.index)
X_test = pd.DataFrame(imputer.fit_transform(X_test),columns=X_test.columns,index=X_test.index)
```

```python
for column in categorical_columns:
    X_train[column]= label_encoder.inverse_transform(X_train[column].astype(int))
    X_test[column]= label_encoder.inverse_transform(X_test[column].astype(int))
```

```python
X_train.fillna(X_train.mode().iloc[0],inplace=True)
X_test.fillna(X_test.mode().iloc[0],inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_columns = X_train.select_dtypes(include=['object']).columns.tolist()

encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(X_train[categorical_columns])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns),index = X_train.index)

X_train_encoded = pd.concat([X_train, one_hot_df], axis=1)
X_train_encoded = X_train_encoded.drop(categorical_columns, axis=1)
```

```python
X_train_encoded
```

```python
one_hot_encoded2 = encoder.fit_transform(X_test[categorical_columns])
one_hot_df2 = pd.DataFrame(one_hot_encoded2, columns=encoder.get_feature_names_out(categorical_columns),index= X_test.index)
X_test_encoded = pd.concat([X_test, one_hot_df2], axis=1)
X_test_encoded = X_test_encoded.drop(categorical_columns, axis=1)
```

```python
X_test_encoded
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear',probability=True,gamma='auto')
svm.fit(X_train_encoded.iloc[:int(0.2*len(X_train_encoded)),:],y_train.iloc[:int(0.2*len(X_train_encoded)),0])
```

```python
svm.fit(X_train_encoded.iloc[int(0.2*len(X_train_encoded)):int(0.4*len(X_train_encoded)),:],y_train.iloc[int(0.2*len(X_train_encoded)):int(0.4*len(X_train_encoded)),0])
```

```python
svm.fit(X_train_encoded.iloc[int(0.4*len(X_train_encoded)):int(0.6*len(X_train_encoded)),:],y_train.iloc[int(0.4*len(X_train_encoded)):int(0.6*len(X_train_encoded)),0])
```

```python
svm.fit(X_train_encoded.iloc[int(0.6*len(X_train_encoded)):int(0.8*len(X_train_encoded)),:],y_train.iloc[int(0.6*len(X_train_encoded)):int(0.8*len(X_train_encoded)),0])
```

```python
svm.fit(X_train_encoded.iloc[int(0.8*len(X_train_encoded)):,:],y_train.iloc[int(0.8*len(X_train_encoded)):,0])
```

```python
prob_pred_svm = svm.predict_proba(X_test_encoded)
```

```python
prob_predictions_svm = pd.DataFrame(index=X_test_encoded.index)
predictions_svm = pd.DataFrame(index=X_test_encoded.index)
for i in range(len(X_test_encoded)):
    prob_predictions_svm.loc[prob_predictions_svm.index[i],'xyz_vaccine']=prob_pred_svm[i][1]
```
